```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

```python
purchase_data_df = pd.DataFrame(purchase_data_df)
purchase_data_df.rename(columns={'SN':'Screen Name'}, inplace=True)
purchase_data_df
```

## Player Count

```python
# * Display the total number of players(x)
```

```python
total_players = len(purchase_data_df['Screen Name'].unique())
purchase_data_df["Total Players"] = total_players
purchase_data_df[["Total Players"]].head(1)
```

# Purchasing Analysis (Total)

```python
# * Run basic calculations to obtain number of unique items, average price, etc.(x)
# * Create a summary data frame to hold the results(x)
# * Optional: give the displayed data cleaner formatting(x)
# * Display the summary data frame(x)
```

```python
total_purchases = purchase_data_df['Purchase ID'].count()
avg_price = ('$' + str(round(purchase_data_df['Price'].mean(),2)))
total_revenue = ("$" + "{:,}".format(purchase_data_df['Price'].sum()))
unique_items = len(purchase_data_df['Item ID'].unique())
purchase_data_df["Number of Unique Items"] = unique_items
purchase_data_df["Average Price"] = avg_price
purchase_data_df["Number of Purchases"] = total_purchases
purchase_data_df['Total Revenue'] = total_revenue
```

```python
purchase_data_df[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue", ]].head(1)
```

## Gender Demographics

```python
# * Percentage and Count of Male Players(x)
# * Percentage and Count of Female Players(x)
# * Percentage and Count of Other / Non-Disclosed(x)
```

```python
player_gender_df = purchase_data_df.loc[:, ["Screen Name", "Age", "Gender"]]
player_gender_df = player_gender_df.drop_duplicates()
gender_total = player_gender_df["Gender"].value_counts()
gender_percent = gender_total / total_players * 100
gender_demographics_df = pd.DataFrame({"Percentage of Players": gender_percent, "Total Count": gender_total})
```

```python
pd.options.display.float_format = '{:,.2f}%'.format
```

```python
gender_demographics_df
```


## Purchasing Analysis (Gender)

```python
# * Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender(x)
# * Create a summary data frame to hold the results(x)
# * Optional: give the displayed data cleaner formatting(x)
# * Display the summary data frame(x)
```

```python
gender_purchase_counts = purchase_data_df.groupby(["Gender"]).count()["Price"]
gender_avg_price = purchase_data_df.groupby(["Gender"]).mean()["Price"]
gender_purchase_total = purchase_data_df.groupby(["Gender"]).sum()["Price"]
avg_total_per_person = gender_purchase_total/gender_demographics_df["Total Count"]
```

```python
pd.options.display.float_format = '${:,.2f}'.format
```

```python
gender_analysis_summary = pd.DataFrame({"Purchase Count": gender_purchase_counts, "Average Purchase Price": gender_avg_price, "Total Purchase Value": gender_purchase_total, "Avg Total Purchase per Person": avg_total_per_person})
gender_analysis_summary
```

## Age Demographics

```python
# * Establish bins for ages(x)
# * Categorize the existing players using the age bins. Hint: use pd.cut()(x)
# * Calculate the numbers and percentages by age group(x)
# * Create a summary data frame to hold the results(x)
# * Optional: round the percentage column to two decimal points(x)
# * Display Age Demographics Table(x)
```

```python
age_bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 999.99]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
player_gender_df["Age Demographic"] = pd.cut(player_gender_df["Age"], age_bins, labels=age_groups, include_lowest=True)
player_gender_df
```

```python
age_total = player_gender_df["Age Demographic"].value_counts()
age_percent = age_total / total_players * 100
```

```python
pd.options.display.float_format = '{:,.2f}%'.format
```

```python
age_demographics_df = pd.DataFrame({"Total Count": age_total, "Percentage of Players": age_percent})
age_demographics_df.sort_index()
```

## Purchasing Analysis (Age)

```python
# * Bin the purchase_data data frame by age(x)
# * Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc.(x)
# * Create a summary data frame to hold the results(x)
# * Optional: give the displayed data cleaner formatting(x)
# * Display the summary data frame()
```

```python
purchase_data_df["Age Demographic"] = pd.cut(purchase_data_df["Age"], age_bins, labels=age_groups, include_lowest=True)
```

```python
age_purchase_counts = purchase_data_df.groupby(["Age Demographic"]).count()["Price"]
age_avg_price = purchase_data_df.groupby(["Age Demographic"]).mean()["Price"]
age_purchase_total = purchase_data_df.groupby(["Age Demographic"]).sum()["Price"]
avg_purchase_per_age = age_purchase_total / age_demographics_df["Total Count"]
```

```python
pd.options.display.float_format = '${:,.2f}'.format
```

```python
age_analysis_summary = pd.DataFrame({"Age Ranges": age_total, "Purchase Count": age_purchase_counts, "Average Purchase Price": age_avg_price, "Total Purchase Value": age_purchase_total, "Avg Total Purchase per Person": avg_purchase_per_age})
age_analysis_summary
```
